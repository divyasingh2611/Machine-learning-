```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= (10,5)
plt.rcParams['figure.dpi']=300
import seaborn as sns
sns.set_theme(style='darkgrid',palette='viridis')
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv("Wholesale customers data.csv")
```

```python
data.head()
```

Since this is grocery products so "Channel"& "Region" are not important for clustering

```python
data.drop(columns=['Channel','Region'],inplace=True,axis=1)
```

```python
data.describe()
```

There are outlier in each columns sicne we'll using 'DBSCAN' it also detect outlier for unlabelled dataset

```python
data.boxplot()
```

```python
sns.pairplot(data)
```

```python
sc = StandardScaler()
x = sc.fit_transform(data)
x
```

# Model Building 

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.5,min_samples=5)
y = db.fit_predict(x)
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1],c= y,cmap='viridis')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,y)
```

# Hyperparameter Tuning

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nn = NearestNeighbors(n_neighbors=3).fit(x)
d,i=nn.kneighbors(x)
```

```python
pd.DataFrame(d)
```

```python
dist = np.sort(d[:,1])
```

```python
plt.plot(dist)
plt.xlabel("Distance")
plt.ylabel("Epsilon value")
```

```python
db = DBSCAN(eps=2,min_samples=3)
y_pred= db.fit_predict(x)
y_pred
```

# Evalute the model

```python
silhouette_score(x,y_pred)
```

```python
plt.scatter(x[:,0],x[:,1],c= y_pred,cmap='viridis')
```

```python
data["cluster"] = y_pred
data
```

```python
data.groupby('cluster').agg("mean")
```
